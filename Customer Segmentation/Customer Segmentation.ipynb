```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
X = df.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss = []
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Within-Cluster-Sum-of-Squares (WCSS) values")
plt.show()
```

#### Thus, the optimal number of cluster is K = 5

```python
kmeansmodel = KMeans(n_clusters=5, init ='k-means++', random_state=0)
```

```python
y_kmeans = kmeansmodel.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0,0], X[y_kmeans == 0,1], s=80, c="red", label="Customer 1")
plt.scatter(X[y_kmeans == 1,0], X[y_kmeans == 1,1], s=80, c="blue", label="Customer 2")
plt.scatter(X[y_kmeans == 2,0], X[y_kmeans == 2,1], s=80, c="yellow", label="Customer 3")
plt.scatter(X[y_kmeans == 3,0], X[y_kmeans == 3,1], s=80, c="cyan", label="Customer 4")
plt.scatter(X[y_kmeans == 4,0], X[y_kmeans == 4,1], s=80, c="black", label="Customer 5")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c="magenta", label="Centroids")
plt.title("Cluster of Customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```

#### Clearly the target customers for the company are the "red cluster" audience, since they are large in number, they earn decent and ofcouse, spend decent amounts. This is the ultimate conclusion we can withdraw from this data visualization, performed by our ML trained model using K-Means Clustering Algorithm.
